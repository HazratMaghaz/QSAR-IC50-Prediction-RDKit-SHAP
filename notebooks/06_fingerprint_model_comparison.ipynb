# 06 — Fingerprint Model Comparison

This notebook summarizes the 987-compound fingerprint QSAR workflow.

It compares multiple molecular fingerprint datasets and machine-learning models for pIC50 prediction.

```python
from pathlib import Path
import pandas as pd

fp_dir = Path("../data/legacy_fingerprints")
files = sorted(fp_dir.glob("*.csv"))
summary = []
for f in files:
    df = pd.read_csv(f)
    summary.append({"fingerprint_file": f.name, "rows": df.shape[0], "columns": df.shape[1]})
summary_df = pd.DataFrame(summary)
display(summary_df)
```

```python
metrics_path = Path("../results/tables/fingerprint_model_metrics.csv")
if metrics_path.exists():
    metrics = pd.read_csv(metrics_path)
    display(metrics.sort_values("r2", ascending=False).head(20))
else:
    print("Metrics not found yet. Run: python scripts/train_fingerprint_qsar_models.py")
```
