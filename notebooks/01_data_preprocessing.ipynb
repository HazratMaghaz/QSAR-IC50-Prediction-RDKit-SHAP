# 01 — QSAR Data Preprocessing

This notebook introduces the QSAR dataset structure and shows how IC50/pIC50 data should be checked before modeling.

The repository contains two data routes:

1. `data/sample_qsar_dataset.csv` — small demo dataset for pipeline testing.
2. `data/legacy_fingerprints/` — real learning fingerprint datasets with 987 compounds and pIC50 values.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("../data")
legacy_dir = DATA_DIR / "legacy_fingerprints"

print("Available data files:")
for p in sorted(DATA_DIR.rglob("*.csv")):
    print("-", p)
```

```python
sample_path = DATA_DIR / "sample_qsar_dataset.csv"
if sample_path.exists():
    sample_df = pd.read_csv(sample_path)
    display(sample_df.head())
    print(sample_df.shape)
else:
    print("Sample dataset not found.")
```

```python
legacy_files = sorted(legacy_dir.glob("*.csv"))
summary = []
for fp in legacy_files:
    df = pd.read_csv(fp)
    summary.append({
        "file": fp.name,
        "rows": df.shape[0],
        "columns": df.shape[1],
        "has_pIC50": "pIC50" in df.columns
    })
summary_df = pd.DataFrame(summary)
display(summary_df)
```
