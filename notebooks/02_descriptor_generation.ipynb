# 02 — Descriptor and Fingerprint Datasets

In classical QSAR, molecules are converted into numerical features such as descriptors or fingerprints.

This project currently uses precomputed fingerprint datasets from a legacy learning workflow. These are stored in `data/legacy_fingerprints/` and already include the `pIC50` target column.

```python
from pathlib import Path
import pandas as pd

fp_dir = Path("../data/legacy_fingerprints")
files = sorted(fp_dir.glob("*.csv"))
print(f"Fingerprint datasets found: {len(files)}")
for f in files:
    print(f.name)
```

```python
# Inspect one high-dimensional fingerprint dataset
selected = fp_dir / "Extended_finterPrinter.csv"
if selected.exists():
    df = pd.read_csv(selected)
    print("Shape:", df.shape)
    display(df.head())
    print("Target summary:")
    display(df["pIC50"].describe())
else:
    print("Selected file not found:", selected)
```

```python
# Feature/target split example
X = df.drop(columns=["pIC50"])
y = df["pIC50"]
print("X shape:", X.shape)
print("y shape:", y.shape)
```
