# 03 — Model Training and Evaluation

This notebook runs the fingerprint QSAR model comparison script.

The script trains multiple ML models across all fingerprint datasets and saves metrics, predictions, figures, and the best model.

```python
# Run from the notebook folder using repo-relative script path
import subprocess, sys

cmd = [sys.executable, "../scripts/train_fingerprint_qsar_models.py"]
result = subprocess.run(cmd, capture_output=True, text=True)
print(result.stdout)
if result.stderr:
    print("STDERR:
", result.stderr)
```

```python
from pathlib import Path
import pandas as pd

metrics_path = Path("../results/tables/fingerprint_model_metrics.csv")
if metrics_path.exists():
    metrics = pd.read_csv(metrics_path)
    display(metrics.sort_values("r2", ascending=False).head(15))
else:
    print("Metrics file not found. Run the training script first.")
```

```python
pred_path = Path("../results/tables/best_model_predictions.csv")
if pred_path.exists():
    preds = pd.read_csv(pred_path)
    display(preds.head())
    print(preds.shape)
else:
    print("Prediction file not found.")
```
